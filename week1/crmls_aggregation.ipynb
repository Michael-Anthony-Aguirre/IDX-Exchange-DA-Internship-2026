```python
import os
import pandas as pd
from pathlib import Path

os.chdir(r"C:\Users\michael\Desktop\IDX Exchange\Files\Week 1\CSVs")
folder = Path(r".")
```

```python
files_sold_names = sorted(folder.glob("CRMLSSold*.csv"))
files_sold_names = [file.name for file in files_sold_names]

files_sold = [pd.read_csv(file, encoding="windows-1252", low_memory = False) for file in files_sold_names]
df_sold = pd.concat(files_sold, ignore_index = True)

sold_concat_before = sum(len(file) for file in files_sold)
sold_concat_after = len(df_sold)

print(f"Sold (before concatenation): {sold_concat_before}")
print(f"Sold (after concatenation): {sold_concat_after}")

# Sold (before concatenation): 655362
# Sold (after concatenation): 655362
```

```python
df_sold_final = df_sold[df_sold.PropertyType == "Residential"]

sold_filter_before = len(df_sold)
sold_filter_after = len(df_sold_final)

print(f"Sold (before filter): {sold_filter_before}")
print(f"Sold (after filter): {sold_filter_after}")

# Sold (before filter): 655362
# Sold (after filter): 438115
```

```python
df_sold_final.to_csv("CRMLSSold.csv", index = False)
```
